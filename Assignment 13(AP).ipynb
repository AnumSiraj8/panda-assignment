```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Salaries - Salaries.csv')
df
```

```python
# Drop rows where EmployeeName and JobTitle are 'Not provided'
df = df[~((df['EmployeeName'] == 'Not provided') & (df['JobTitle'] == 'Not provided'))]
df
```

```python
# Replace 'Not Provided' with 0 in BasePay OvertimePay OtherPay Benefits
df.BasePay.replace('Not Provided',0).astype(float)
df.OvertimePay.replace('Not Provided',0).astype(float)
df.OtherPay.replace('Not Provided',0).astype(float)
df.Benefits.replace('Not Provided',0).astype(float)
```

```python
# Change BasePay OvertimePay OtherPay Benefits type to float
df.BasePay.astype(float)
df.OvertimePay.astype(float)
df.OtherPay.astype(float)
df.Benefits.astype(float)
```

```python
# Display all employees name and basepay columns who have a BasePay greater than 200,000.
highly_paid = df[df['BasePay'] > 200000][['EmployeeName', 'BasePay']]
print(highly_paid)
```

```python
# Find all employees whose JobTitle contains the word “Manager” (case = False)
managers = df[df['JobTitle'].str.contains('Manager', case=False, na=False)]
print(managers[['EmployeeName', 'JobTitle']])
```

```python
# Retrieve only the columns: EmployeeName, JobTitle, and TotalPay.
retrieve_columns = df[['EmployeeName', 'JobTitle', 'TotalPay']]
print(retrieve_columns)
```

```python
# Find count of employees who have zero OvertimePay.
count_zero_overtime = (df['OvertimePay'] == 0).sum()
print(count_zero_overtime)
```

```python
# Calculate the TotalPay for each JobTitle where totalpay > 300000
new = df[df['TotalPay'] > 300000]
new_results = new.groupby('JobTitle')['TotalPay'].sum().reset_index()
print(new_results)
```

```python
# Find the top 5 JobTitles with the highest average TotalPayBenefits show in descending order
avgerage_pay = df.groupby('JobTitle')['TotalPayBenefits'].mean().reset_index()
top_five = avgerage_pay.sort_values(by='TotalPayBenefits', ascending=False).head(5)
print(top_five)
```
